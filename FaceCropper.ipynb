```python
import os

import cv2

image_path = "Data/celebA/img_align_celeba_png"


def save_faces(cascade, imgname):
    dimension = (90,90)
    img = cv2.imread(os.path.join(image_path, imgname))
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    for i, face in enumerate(cascade.detectMultiScale(gray,1.1,4)):
        x, y, w, h = face
        sub_face = img[y:y + h, x:x + w]
        sub_face = cv2.resize(sub_face, dimension, interpolation = cv2.INTER_AREA)
        #print('i:{}'.format(i))
        if (h>90 and w>90):
            cv2.imwrite(os.path.join("Data/celebA/", "{}_{}.jpg".format(imgname.removesuffix('.png'), i)), sub_face)
        #print(os.path.join("faces/", "{}_{}.jpg".format(imgname.removesuffix('.png'), i)))
        #print('loop end')

if __name__ == '__main__':
    face_cascade = "haarcascade_frontalface_alt2.xml"
    cascade = cv2.CascadeClassifier(face_cascade)
    number_of_images_needed = 1000
    # Iterate through files
    for f in [f for f in os.listdir(image_path)[:number_of_images_needed] if os.path.isfile(os.path.join(image_path, f))]:
        save_faces(cascade, f)
        print(f.removesuffix('.png'))
```
